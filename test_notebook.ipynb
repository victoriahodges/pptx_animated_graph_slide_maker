```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y0 = np.sin(x)
y1 = np.cos(x)
```

```python
plt.plot(x, y0)
plt.title("Image 1")
plt.show()
plt.plot(x, y1)
plt.title("Image 2")
plt.show()
```
